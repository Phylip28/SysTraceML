```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "data/Training_Data_Master/UTD-0001.txt"
```

```python
with open(file, "r") as f:
    syscalls = list(map(int, f.read().split()))
```

```python
df = pd.DataFrame(syscalls, columns=["syscall_id"])
```

```python
print(df.head())
print(df.describe())
print(df["syscall_id"].value_counts().sort_index())
```

```python
syscalls_ids = np.array([1,"otros",252])
frequencies = np.array([539,13,267])

plt.figure(figsize=(10, 5))
plt.bar(syscalls_ids, frequencies, width=0.5)
plt.xticks(syscalls_ids,rotation=45)
plt.xlabel("syscall id")
plt.ylabel("Frequency")
plt.title("Syscall Frequency")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.pie(
    labels=syscalls_ids,
    x=frequencies,
    autopct="%1.1f%%",
)
plt.title("Syscall Frequency")
plt.show()
```
